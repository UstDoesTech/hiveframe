# HiveFrame Notebook Example

This notebook demonstrates the Jupyter-style interface for HiveFrame with Python and SQL support.

```python
from hiveframe import HiveDataFrame

# Create sample sales data
sales_data = [
    {"product": "Laptop", "price": 1200, "quantity": 5, "category": "Electronics"},
    {"product": "Mouse", "price": 25, "quantity": 50, "category": "Electronics"},
    {"product": "Desk", "price": 300, "quantity": 10, "category": "Furniture"},
    {"product": "Chair", "price": 150, "quantity": 20, "category": "Furniture"},
    {"product": "Monitor", "price": 400, "quantity": 15, "category": "Electronics"}
]

# Create DataFrame
df = HiveDataFrame(sales_data)

# Register table for SQL queries
_sql_context.register_table("sales", df)

print("Sales data loaded!")
df.show()
```

```python
SELECT category, SUM(price * quantity) as total_revenue
FROM sales
GROUP BY category
ORDER BY total_revenue DESC
```

```python
SELECT product, price, quantity, (price * quantity) as revenue
FROM sales
WHERE category = 'Electronics'
ORDER BY revenue DESC
```

```python
# You can also use Python for analysis
electronics = df.filter(lambda row: row['category'] == 'Electronics')
print(f"\nTotal Electronics Products: {electronics.count()}")
electronics.show()
```
